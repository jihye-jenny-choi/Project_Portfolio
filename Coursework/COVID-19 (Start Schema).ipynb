* Creating a merge table which has values from cases, vaccine, people_vaccinated and testing. We perform a FULL OUTER JOIN to create merge table to avoid loss of data.

```python
%%sql
DROP TABLE IF EXISTS merge;

CREATE TABLE merge AS (
SELECT t.*, c.province_state, c.country_region, c.last_update, c.lat, c.long, 
       c.confirmed, c.deaths, c.recovered, c.active, c.fips as fips1, c.incident_rate, 
       c.total_test_results, c.people_hospitalized, c.case_fatality_ratio, c.uid, c.iso3, 
       c.testing_rate, c.hospitalization_rate, c.primary_key as primary_key1, 
       v.province_state as province_state1, v.date as date1, v.vaccine_type, v.fips as fips2, 
       v.country_region as country_region1, v.lat as lat1, v.long as long1, v.doses_alloc, v.doses_shipped, 
       v.doses_admin, v.stage_one_doses, v.stage_two_doses, v.combined_key, v.primary_key as primary_key2, p.fips as fips3,
       p.province_state as province_state2, p.country_region as country_region2, p.date as date2, p.lat as lat2, 
       p.long as long2, p.combined_key as combined_key1, p.people_fully_vaccinated, p.people_partially_vaccinated, 
       p.primary_key as primary_key3
from testing t
FULL OUTER JOIN cases c on t.state = c.province_state and t.date = c.last_update
FULL OUTER JOIN vaccine v on t.state = v.province_state and t.date = v.date
FULL OUTER JOIN people_vaccinated p on t.state = p.province_state and t.date = p.date
)
```

#### Create state_dim table

* We create state_dim dimension table by selecting different columns from the merged table
* Checking all the unique values of states

```python
%%sql
SELECT DISTINCT fips, state
FROM merge
UNION
SELECT DISTINCT fips1, province_state
FROM merge
UNION
SELECT DISTINCT fips2, province_state1
FROM merge
UNION
SELECT DISTINCT fips3, province_state2
FROM merge
```

* Creating the state_dim table here 

```python
%%sql
--ALTER TABLE merge DROP CONSTRAINT fk_state;
DROP TABLE IF EXISTS state_dim;

CREATE TABLE state_dim (
    state_key SERIAL PRIMARY KEY,
    fips varchar(8),
    state_US VARCHAR(64),
    state_code_US CHAR(2),
    lat FLOAT,
    long FLOAT
);
```

* Inserting values into the state_dim dimension table

```python
%%sql
INSERT INTO state_dim (fips,state_US)
SELECT DISTINCT fips, state
FROM merge
UNION
SELECT DISTINCT fips1, province_state
FROM merge
UNION
SELECT DISTINCT fips2, province_state1
FROM merge
UNION
SELECT DISTINCT fips3, province_state2
FROM merge
order by state
```

* Creating a new dimension table day_dim

```python
%%sql
--ALTER TABLE merge DROP CONSTRAINT fk_day;
DROP TABLE IF EXISTS day_dim;

CREATE TABLE day_dim (
    day_key SERIAL PRIMARY KEY,
    date date,
    year INTEGER,
    month_of_year_str VARCHAR(12),
    month_of_year INTEGER,
    day_of_month INTEGER,
    day_of_week_str CHAR(9),
    day_of_week INTEGER,
    is_weekend BOOLEAN,
    is_weekday BOOLEAN,
    quarter_of_year INTEGER
);
```

* Making elaborative day_dim table by inserting yearwise, querterwise, daywise information in different columns

```python
%%sql
INSERT INTO day_dim (date, year, month_of_year_str, month_of_year, day_of_month, 
                  day_of_week_str, day_of_week, is_weekend, is_weekday, quarter_of_year)
SELECT date AS date, 
    CAST(TO_CHAR(date, 'YYYY') AS INTEGER) AS year,
    TO_CHAR(date, 'Month') AS month_of_year_str,
    CAST(TO_CHAR(date, 'MM') AS INTEGER) AS month_of_year,
    CAST(TO_CHAR(date, 'DD') AS INTEGER) AS day_of_month,
    TO_CHAR(date, 'Day') AS day_of_week_str,
    CAST(TO_CHAR(date, 'D') AS INTEGER) AS day_of_week,
    CASE WHEN CAST(TO_CHAR(date, 'D') AS INTEGER) IN (1, 7) 
        THEN TRUE
        ELSE FALSE
    END AS is_weekend,
    CASE WHEN CAST(TO_CHAR(date, 'D') AS INTEGER) NOT IN (1, 7) 
        THEN TRUE
        ELSE FALSE
    END AS is_weekday,
    CAST(TO_CHAR(date, 'Q') AS INTEGER) AS quarter_of_year
FROM view3
```
